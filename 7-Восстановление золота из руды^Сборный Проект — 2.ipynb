Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Используем данные с параметрами добычи и очистки.

**Технологический процесс**
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- Xanthate — ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи


**Наименование признаков**

Наименование признаков должно быть такое:

[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag


**Возможные значения для блока [этап]:**
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики
- Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

Напишем функцию для вывода основных статистик набораа данных.

```python
def dataframe_information(data):
    """Функция для вывода основных статистик набора данных"""
    df_data = []
    df_cols = ['name', 'object', 'na', 'zero', 'rate', 'unique', 'neg',
               'mean', 'std', 'min', 'max', 'outliners', 'duplicates']
    # перебираем столбцы в наборе данных
    for column_name in data.columns:
        column_negative_values, column_mean, column_std, column_min, column_max, outliers, duplicates_sum = \
            None, None, None, None, None, None, None
        # считаем характеристики
        column_type = data[column_name].dtypes
        column_na_values = data[column_name].isna().sum()
        column_zero_values = data[column_name][data[column_name] == 0].count()
        column_na_zero_rate = ((column_na_values + column_zero_values) / data.shape[0]) * 100
        column_unique_values = len(data[column_name].unique())
        duplicates_sum = len(data.loc[data.duplicated() > 0, column_name])
        if data[column_name].dtype != 'object':
            column_negative_values = data[column_name][data[column_name] < 0].count()
            column_mean = data[column_name].mean()
            column_std = data[column_name].std()
            column_min = data[column_name].min()
            column_max = data[column_name].max()
            q1 = data[column_name].quantile(0.25)
            q3 = data[column_name].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - 1.5 * iqr
            upper_bound = q3 + 1.5 * iqr
            outliers = len(data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)])

        # собираем показатели по каждому полю
        df_data.append([column_name, column_type, column_na_values,
                        column_zero_values, column_na_zero_rate, column_unique_values,
                        column_negative_values, column_mean, column_std,
                        column_min, column_max, outliers, duplicates_sum])

    # формируем набор данных
    df_res = pd.DataFrame(data=df_data, columns=df_cols)
    s_isna = data.isna().any(axis=1).sum()
    s_null = (data == 0).any(axis=1).sum()
    s_isna_and_null = ((data == 0) | (data.isna())).any(axis=1).sum()
    

    return (
        print(f'_____________________________________________________________________________\n\nДатафрейм {df}:'),
        display(df_res),
        print(f'Всего пропусков - {data.isna().sum().sum()}\nСтрок с пропусками - {s_isna}\n'
              f'Процент от всего датафрейма - {s_isna / len(data) * 100:.2f}%\n\n'
              f'Всего нулевых значений - {(data == 0).sum().sum()}\nСтрок с нулями - {s_null}\n'
              f'Процент от всего датафрейма - {s_null / len(data) * 100:.2f}%\n\n'
              f'Строки с пропусками и нулями - {s_isna_and_null}\n'
              f'Строки с пропусками и нулями одновременно - {s_null + s_isna - s_isna_and_null}\n'
              f'Процент от всего датафрейма - {(s_null + s_isna - s_isna_and_null) / len(data) * 100:.2f}%'
        ),
    )


# Применим функцию на датафреймы
for df in dataframes_names:
    dataframe_information(eval(df))
```

Для оценки риска мельтиколлинеарности при использовании модели линейной регрессии, посмотрим корреляцию между колонками. Но так как колонок очень много, просто сформируем список колонок, корреляция которых с другими столбцами больше 0.8.

```python
# Для обзора данных апишем функцию для нахождения колонок с максимальной корреляцией
def get_columns_with_max_correlation(data) -> list[dict[str: int]]:
    max_corr_list = []
    # чтобы не было ошибки, ограничим нужные колонки только числовыми
    columns = data.select_dtypes(include='number').columns
    for indx in range(len(columns)):
        correlation_temp = ['', 0]
        for indx2 in range(len(columns)):
            if columns[indx] != columns[indx2]:
                correlation = data[columns[indx]].corr(data[columns[indx2]])
                if correlation_temp[1] < abs(correlation):
                    correlation_temp = [columns[indx2], correlation]
        if correlation_temp[1] > 0.8:
            max_corr_list.append(columns[indx])
    return max_corr_list


# Выведем на экран списки
for df in dataframes_names:
    print(f'Список сильно кореллирующих колонок для {df}')
    display(get_columns_with_max_correlation(eval(df)))
    print('___________________________________________________')
    
```

### Проверка правильности расчета эффективности обогащения.

Эффективность обогащения рассчитывается по формуле: \
((С х (F - T)) / (F x (C - T))) x 100 

где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Нам нужно проверить, правильно ли рассчитан целевой признак rougher.output.recovery в обучающей выборке. Найдем МАЕ между нашими расчетами и значением признака.

```python
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']

recovery = ((C * (F - T)) / (F * (C - T))) * 100 

mean_absolute_error(df_train['rougher.output.recovery'], recovery)
```

```python
df_full[['secondary_cleaner.output.tail_au', 'final.output.tail_au']].head()
```

С точки зрения чисел с плавающей запятой, такое значение МАЕ можно считать пренебрежимо малым и практически равным нулю, поэтому можно уверенно сказать, что целевой признак рассчитан корректно.

### Расчет итоговой sMAPE

Напишем функцию для вычисления итоговой sMAPE. 

Формула для ее вычисления выглядит так: 

sMAPE = (1 / n) * Σ(|A - F| / (|A| + |F|) / 2) * 100

Где:

A - фактическое значение \
F - прогнозируемое или предсказанное значение \
n - количество точек данных. \
Показатель sMAPE рассчитывает абсолютную процентную разницу между каждым фактическим и прогнозным значением, суммирует их и берет среднее значение. От традиционного MAPE он отличается тем, что в знаменателе используется среднее значение фактического и прогнозного значений, что обеспечивает симметрию.

Полученное значение sMAPE выражается в процентах, где меньшие значения указывают на более высокую точность.

Так как нам нужно спрогнозировать сразу две величины (rougher.output.recovery и final.output.recovery), формула для вычисления итоговой sMAPE выглядит так:

final_sMAPE = 25% * sMAPE(rougher.output.recovery) + 75% * sMAPE(final.output.recovery)

```python
# Напишем функцию для расчета sMAPE
def smape(true, pred):
    return (abs(true - pred) / ((abs(true) + abs(pred)) / 2)).mean() * 100

# Напишем функцию для расчета итоговой sMAPE
def final_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

#### Решающее дерево

```python
%%time
# Т.к. нам нужно чтобы качество модели измерялось не метрикой R2, а sMAPE
# создадим собственный scorer
smape_scorer = make_scorer(smape)

best_model = None
best_depth = 0
best_result = 200
best_smape_rougher = 200
best_smape_final = 200

# Возьмем максимульную глубину дерева - 15
for depth in range(1, 15):
    model = DecisionTreeRegressor(random_state=RANDOM_STATE, max_depth=depth)
    smape_rougher = cross_val_score(
        model, features_train_rougher, target_train_rougher, scoring=smape_scorer, cv=5
    ).mean()
    smape_final = cross_val_score(
        model, features_train_final, target_train_final, scoring=smape_scorer, cv=5
    ).mean()    
    result = final_smape(smape_rougher, smape_final)
    if result < best_result:
        best_model = model
        best_result = result
        best_depth = depth
        best_smape_rougher = smape_rougher
        best_smape_final = smape_final

print(f'sMAPE наилучшей модели для предсказания rougher.output.recovery: {best_smape_rougher:.2f}')
print(f'sMAPE наилучшей модели для предсказания final.output.recovery: {best_smape_final:.2f}')
print(f'Финальный sMAPE наилучшей модели равен: {best_result:.2f}')
print(f'Оптимальная глубина дерева - {best_depth}')
```

#### Случайный лес

```python
%%time
# Wall time: 5min 48s
best_model_trees = None
best_depth = 0
best_est = 0
best_result = float('inf')
best_smape_rougher = float('inf')
best_smape_final = float('inf')

for est in range(5, 46, 15):
    for depth in range(2, 19, 4):
        model = RandomForestRegressor(
            random_state=RANDOM_STATE, n_estimators=est, max_depth=depth
        )
        smape_rougher = cross_val_score(
            model, features_train_rougher, target_train_rougher, scoring=smape_scorer, cv=3
        ).mean()
        smape_final = cross_val_score(
            model, features_train_final, target_train_final, scoring=smape_scorer, cv=5
        ).mean()
        result = final_smape(smape_rougher, smape_final)
        if result < best_result:
            best_model_trees = model
            best_result = result
            best_depth = depth
            best_est = est
            best_smape_rougher = smape_rougher
            best_smape_final = smape_final

print(f'sMAPE наилучшей модели для предсказания rougher.output.recovery: {best_smape_rougher:.2f}')
print(f'sMAPE наилучшей модели для предсказания final.output.recovery: {best_smape_final:.2f}')
print(f'Финальный sMAPE наилучшей модели равен: {best_result:.2f}')
print(f'Оптимальная глубина дерева - {best_depth}')
print(f'Лучшее количество деревьев - {best_est}')
```
